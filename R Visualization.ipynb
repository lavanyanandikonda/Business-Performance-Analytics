```r
library(ggplot2)
library(dplyr)

# Create a new data frame with the year, total sales, and total profit
superstore_sales_summary <- data.frame(
  year = c(2014, 2015, 2016, 2017, 2018),
  total_sales = c(219,234,239,254,269),
  total_profit = c(9,11,12,14,16)
)

# Create a line chart of the total sales and profit for each year
ggplot(superstore_sales_summary, aes(x = year, y = total_sales)) +
  geom_line() +
  geom_point() +
  xlab("Year") +
  ylab("Total Sales") +
  ggtitle("Total Sales and Profit by Year")

# Create a bar chart of the total profit for each year
ggplot(superstore_sales_summary, aes(x = year, y = total_profit, fill = year)) +
  geom_bar(stat = "sum") +
  xlab("Year") +
  ylab("Total Profit") +
  ggtitle("Total Profit by Year")
```

```r
library(ggplot2)
library(dplyr)
superstore_sales <- read.csv("/content/Dataset.csv")

# Create a new data frame with the region, total sales, and total profit
superstore_sales_summary <- superstore_sales %>%
  group_by(region) %>%
  summarise(total_sales = sum(sales), total_profit = sum(profit))

# Order the data frame by total profit in descending order
superstore_sales_summary <- superstore_sales_summary %>%
  arrange(desc(total_profit))

# Create a bar chart of the total sales and profit for each region
ggplot(superstore_sales_summary, aes(x = region, y = total_profit, fill = region)) +
  geom_bar(stat = "sum") +
  xlab("Region") +
  ylab("Total Profit") +
  ggtitle("Total Profit by Region") +
  coord_flip() +
  theme(axis.text.x = element_text(angle = 180, hjust = 1))
```

```r
library(ggplot2)
library(dplyr)
superstore_sales <- read.csv("/content/Dataset.csv")

# Calculate the total sales, total profits, and profit margin for each region
superstore_profit_margin <- superstore_sales %>%
  group_by(region) %>%
  summarize(profit_margin = (sum(profit) / sum(sales)) * 100)

# Order the data frame by profit margin in descending order
superstore_profit_margin <- superstore_profit_margin %>%
  arrange(desc(profit_margin))

# Create a bar chart of the profit margin for each region
ggplot(superstore_profit_margin, aes(x = region, y = profit_margin)) +
  geom_bar(stat = "identity", position = "dodge") +
  xlab("Region") +
  ylab("Profit Margin (%)") +
  ggtitle("Profit Margin by Region in Superstore Sales Data") +
  coord_flip() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))

# Save the plot to a file
ggsave("profit_margin_by_region.png", width = 8, height = 6)
```

```r
library(ggplot2)
library(dplyr)
superstore_sales <- read.csv("/content/Dataset.csv")

# Calculate the total sales, total profits, and profit margin for each state
superstore_profit_margin <- superstore_sales %>%
  group_by(state) %>%
  summarize(Total_Sales = sum(sales), Total_Profits = sum(profit), profit_margin = (sum(profit) / sum(sales)) * 100)

# Order the data frame by profit margin in descending order
superstore_profit_margin <- superstore_profit_margin %>%
  arrange(desc(profit_margin))

# Limit the data frame to the top 10 states
superstore_profit_margin <- superstore_profit_margin %>%
  head(10)

# Create a bar chart of the profit margin for each state
ggplot(superstore_profit_margin, aes(x = state, y = profit_margin)) +
  geom_bar(stat = "identity", position = "dodge") +
  xlab("State") +
  ylab("Profit Margin (%)") +
  ggtitle("Top 10 States with Highest Profit Margin in Superstore Sales Data") +
  coord_flip() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
ggsave("top_10_states_with_highest_profit_margin.png", width = 8, height = 6)
```

```r
library(ggplot2)
library(dplyr)
superstore_sales <- read.csv("/content/Dataset.csv")

# Calculate the total sales, total profits, and profit margin for each city
superstore_profit_margin <- superstore_sales %>%
  group_by(city) %>%
  summarize(Total_Sales = sum(sales), Total_Profits = sum(profit), profit_margin = (sum(profit) / sum(sales)) * 100, 2)

# Order the data frame by profit margin in descending order
superstore_profit_margin <- superstore_profit_margin %>%
  arrange(desc(profit_margin))

# Limit the data frame to the top 10 cities
superstore_profit_margin <- superstore_profit_margin %>%
  head(10)

# Create a bar chart of the profit margin for each city
ggplot(superstore_profit_margin, aes(x = city, y = profit_margin)) +
  geom_bar(stat = "identity", position = "dodge") +
  xlab("City") +
  ylab("Profit Margin (%)") +
  ggtitle("Top 10 Cities with Highest Profit Margin in Superstore Sales Data") +
  coord_flip() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
ggsave("top_10_cities_with_highest_profit_margin.png",  width = 8, height = 6)
```

```r
library(ggplot2)
library(dplyr)
superstore_sales <- read.csv("/content/Dataset.csv")

# Calculate the total sales and total profit for each region and category
superstore_sales_summary <- superstore_sales %>%
  group_by(region, category) %>%
  summarize(total_sales = sum(sales), total_profit = sum(profit))

# Order the data frame by total profit in descending order
superstore_sales_summary <- superstore_sales_summary %>%
  arrange(desc(total_profit))

# Create a bar chart of the total profit for each region and category
ggplot(superstore_sales_summary, aes(x = category, y = total_profit, color = region, fill = region)) +
  geom_bar(stat = "sum", position = "dodge") +
  xlab("Category") +
  ylab("Total Profit") +
  ggtitle("Total Profit by Region and Category in Superstore Sales Data") +
  facet_wrap(~region, scales = "free")
ggsave("total_profit_by_region_and_category.png", width = 10, height = 8)
```
